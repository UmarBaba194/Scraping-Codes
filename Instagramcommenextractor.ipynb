```python
import langid
import re
import csv

def clean_text(text):
    # Remove emojis using a regex pattern
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # Emoticons
                               u"\U0001F300-\U0001F5FF"  # Symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # Transport & map symbols
                               u"\U0001F700-\U0001F77F"  # Alchemical symbols
                               u"\U0001F780-\U0001F7FF"  # Geometric shapes
                               "]+", flags=re.UNICODE)
    text = emoji_pattern.sub(r'', text)

    # Remove unnecessary spaces
    text = ' '.join(text.split())

    # Remove punctuation and special characters
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)

    # Remove hyperlinks
    text = re.sub(r'http\S+', '', text)

    return text

def extract_codeswitched_comments_from_csv(csv_file_path):
    with open(csv_file_path, 'r', encoding='utf-8') as file:
        csv_reader = csv.reader(file)
        header = next(csv_reader, None)

        if header:
            text_column_index = header.index('text') if 'text' in header else 0
        else:
            text_column_index = 0

        codeswitched_comments = []

        for row in csv_reader:
            text = row[text_column_index]
            cleaned_text = clean_text(text)

            # Use langid to identify the language of each sentence
            sentences = cleaned_text.split('\n')
            codeswitched_sentences = [sentence for sentence in sentences if langid.classify(sentence)[0] in ['en', 'ha']]

            # Join the selected comments into a single string
            extracted_codeswitched_comments = '\n'.join(codeswitched_sentences)

            codeswitched_comments.append(extracted_codeswitched_comments)

    return codeswitched_comments

def save_to_csv(output_file_path, comments):
    with open(output_file_path, 'w', encoding='utf-8', newline='') as csvfile:
        csv_writer = csv.writer(csvfile)
        csv_writer.writerow(['codeswitched_comments'])  # Header

        for comment in comments:
            csv_writer.writerow([comment])

# Example usage:
csv_file_path = "extracted_comments.csv"
output_csv_file_path = "output_codeswitched_comments.csv"

codeswitched_comments = extract_codeswitched_comments_from_csv(csv_file_path)
save_to_csv(output_csv_file_path, codeswitched_comments)

print(f"Output saved to {output_csv_file_path}")
```

```python
import langid
import re
import csv
import pandas as pd

def clean_text(text):
    # Remove emojis using a regex pattern
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # Emoticons
                               u"\U0001F300-\U0001F5FF"  # Symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # Transport & map symbols
                               u"\U0001F700-\U0001F77F"  # Alchemical symbols
                               u"\U0001F780-\U0001F7FF"  # Geometric shapes
                               "]+", flags=re.UNICODE)
    text = emoji_pattern.sub(r'', text)

    # Remove unnecessary spaces
    text = ' '.join(text.split())

    # Remove punctuation and special characters
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)

    # Remove hyperlinks
    text = re.sub(r'http\S+', '', text)

    return text

def extract_codeswitched_comments_from_csv(csv_file_path):
    df = pd.read_csv(csv_file_path)
    df.dropna(subset=['Extracted Comments'], inplace=True)  # Drop rows with nan values in the 'text' column

    codeswitched_comments = []

    for text in df['Extracted Comments']:
        cleaned_text = clean_text(text)

        # Use langid to identify the language of each sentence
        sentences = cleaned_text.split('\n')
        codeswitched_sentences = [sentence for sentence in sentences if langid.classify(sentence)[0] in ['en', 'ha']]

        # Join the selected comments into a single string
        extracted_codeswitched_comments = '\n'.join(codeswitched_sentences)

        codeswitched_comments.append(extracted_codeswitched_comments)

    return codeswitched_comments

def save_to_csv(output_file_path, comments):
    with open(output_file_path, 'w', encoding='utf-8', newline='') as csvfile:
        csv_writer = csv.writer(csvfile)
        csv_writer.writerow(['Extracted Comments'])  # Header

        for comment in comments:
            csv_writer.writerow([comment])

# Example usage:
csv_file_path = "extracted_comments.csv"
output_csv_file_path = "output_codeswitched_comments.csv"

codeswitched_comments = extract_codeswitched_comments_from_csv(csv_file_path)
save_to_csv(output_csv_file_path, codeswitched_comments)

print(f"Output saved to {output_csv_file_path}")
```
