```python
import re
import langid
import nltk
from nltk import sent_tokenize

# Download the punkt tokenizer if not already downloaded
#nltk.download('punkt')

def remove_emojis(text):
    # Remove emojis using a regex pattern
    emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # Emoticons
        u"\U0001F300-\U0001F5FF"  # Symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # Transport & map symbols
        u"\U0001F700-\U0001F77F"  # Alchemical symbols
        u"\U0001F780-\U0001F7FF"  # Geometric shapes
        u"\U0001F800-\U0001F8FF"  # Extended emoticons
        u"\U0001F900-\U0001F9FF"  # Supplemental symbols and pictographs
        u"\U0001FA00-\U0001FA6F"  # Extended symbols for ligatures, etc.
        u"\U0001FA70-\U0001FAFF"  # Additional emoticons
        "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)

def is_code_switched(sentence):
    # Use langid to detect the language of the sentence
    lang, _ = langid.classify(sentence)
    # Check if the detected language is not English
    return lang != 'en'

def extract_code_switched_comments(text):
    # Remove emojis
    text_no_emojis = remove_emojis(text)

    # Tokenize sentences
    sentences = sent_tokenize(text_no_emojis)

    # Extract the first sentence after the numeric markers if it's code-switched
    code_switched_comments = []
    for i in range(len(sentences)):
        # Check if the sentence starts with a numeric marker
        if re.match(r'\b\d+[a-zA-Z]*\b', sentences[i]):
            comment_start = i + 3
            # Skip empty sentences
            while comment_start < len(sentences) and not sentences[comment_start].strip():
                comment_start += 1
            # Check if the sentence contains "profile picture"
            if comment_start < len(sentences) and 'profile picture' not in sentences[comment_start].lower():
                # Add the non-empty and non-"profile picture" sentence to the comments list if it's code-switched
                if is_code_switched(sentences[comment_start]):
                    code_switched_comments.append(sentences[comment_start].strip())

    # Join the selected code-switched comments into a single string
    extracted_comments = '\n'.join(code_switched_comments)

    return extracted_comments

# Load extracted comments from the file
with open('comments2.txt', 'r', encoding='utf-8') as file:
    extracted_comments_text = file.read()

# Extract code-switched comments
extracted_code_switched_comments = extract_code_switched_comments(extracted_comments_text)

# Save the extracted code-switched comments to a new text file
with open('extracted_comments2.txt', 'w', encoding='utf-8') as file:
    file.write(extracted_code_switched_comments)
```

```python
import googleapiclient.discovery
import pandas as pd

def fetch_comments(video_id, max_results=500):
    comments = []
    next_page_token = None

    while len(comments) < max_results:
        request = youtube.commentThreads().list(
            part="snippet",
            videoId=video_id,
            maxResults=min(100, max_results - len(comments)),
            pageToken=next_page_token
        )
        response = request.execute()

        for item in response['items']:
            comment = item['snippet']['topLevelComment']['snippet']
            comments.append([
                comment['authorDisplayName'],
                comment['publishedAt'],
                comment['updatedAt'],
                comment['likeCount'],
                comment['textDisplay']
            ])

        next_page_token = response.get('nextPageToken')

        if not next_page_token:
            break

    return comments

api_service_name = "youtube"
api_version = "v3"
DEVELOPER_KEY = "AIzaSyBFjmeYeFogH-ek1v7YWohypPgc3REsTC4"

youtube = googleapiclient.discovery.build(
    api_service_name, api_version, developerKey=DEVELOPER_KEY)

video_id = "0Uibq7UTMXU"
max_results = 1000

comments = fetch_comments(video_id, max_results)

df = pd.DataFrame(comments, columns=['author', 'published_at', 'updated_at', 'like_count', 'text'])

df.to_csv('youtube_comments2.csv', index=False)
df.head(10)
```
