```python
import pandas as pd
import re

# Specify your input CSV file
input_csv_file = "test16.csv"

# Specify your output CSV file
output_csv_file = "your_output_file16.csv"

# Read the CSV file into a DataFrame using semicolon as the delimiter
df = pd.read_csv(input_csv_file, delimiter=';')

# Print the existing columns in your DataFrame
print("Existing columns:", df.columns)

# Define a regular expression pattern to match "\\uXXXX" where XXXX is a hexadecimal number
pattern = re.compile(r'\\u[0-9a-fA-F]{4}')

# Check if the "Comments" column exists in your DataFrame
if ' Comments' in df.columns:
    # Apply the regular expression to the "Comments" column and replace the matched patterns
    df[' Comments'] = df[' Comments'].replace(pattern, '', regex=True)

    # Save the modified DataFrame to a new CSV file
    df.to_csv(output_csv_file, index=False, sep=';')
    print("Modification complete. Output saved to", output_csv_file)
else:
    print("The 'Comments' column does not exist in the DataFrame.")
```

```python
import langid

def detect_language(text):
    # Detect language using langid
    lang, _ = langid.classify(text)
    return lang

def find_mixed_language_sentences(file_path):
    mixed_language_sentences = []

    with open(file_path, 'r', encoding='utf-8') as text_file:
        lines = text_file.readlines()

    for line in lines:
        line = line.strip()
        if line:
            # Check if the sentence has a mixture of languages
            languages = set(detect_language(word) for word in line.split())
            if len(languages) > 1:
                mixed_language_sentences.append(line)

    return mixed_language_sentences

# Example usage
input_text_file = 'Filenaname16output.csv'
output_mixed_language_sentences_file = 'mixed_language_sentences15.txt'

mixed_language_sentences = find_mixed_language_sentences(input_text_file)

# Save mixed language sentences to a new file
with open(output_mixed_language_sentences_file, 'w', encoding='utf-8') as mixed_language_file:
    mixed_language_file.write('\n'.join(mixed_language_sentences))

print(f'Mixed language sentences detected from {input_text_file}. Saved to {output_mixed_language_sentences_file}.')
```
